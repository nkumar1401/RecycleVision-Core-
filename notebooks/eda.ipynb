```python
import os
import matplotlib.pyplot as plt
import seaborn as snsP
import numpy as np
from PIL import Image

# Path to your Kaggle dataset
DATA_DIR = '../garbage_classification'

# 1. Analyze Class Distribution
classes = os.listdir(DATA_DIR)
counts = {cls: len(os.listdir(os.path.join(DATA_DIR, cls))) for cls in classes}

plt.figure(figsize=(12, 6))
sns.barplot(x=list(counts.keys()), y=list(counts.values()))
plt.title("Number of Images per Category")
plt.xlabel("Waste Type")
plt.ylabel("Count")
plt.show()

# 2. Visualize Sample Images
fig, axes = plt.subplots(1, len(classes), figsize=(20, 5))
for i, cls in enumerate(classes):
    img_name = os.listdir(os.path.join(DATA_DIR, cls))[0]
    img_path = os.path.join(DATA_DIR, cls, img_name)
    img = Image.open(img_path)
    axes[i].imshow(img)
    axes[i].set_title(cls)
    axes[i].axis('off')
plt.suptitle("Sample Waste Categories")
plt.show()

# Calculating average image resolution to ensure consistency
resolutions = []
for cls in classes:
    for img_name in os.listdir(os.path.join(DATA_DIR, cls))[:10]:
        with Image.open(os.path.join(DATA_DIR, cls, img_name)) as img:
            resolutions.append(img.size)

avg_res = np.mean(resolutions, axis=0)
print(f"Average Image Resolution: {avg_res[0]:.0f}x{avg_res[1]:.0f}")
```
